# Feature Scaling

- Feature scaling is a technique used to make independent features present in data in a fixed range.

## 1. Standardization

<img src="https://miro.medium.com/v2/resize:fit:1284/1*IkPjUeQPMDL6ernTiEeDsA.png" >

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('bank-full.csv',sep=';',usecols=['age','balance','day','duration','y'])
data.head()
```

```python
data['y'] = data['y'].replace({'no':0, 'yes':1})
```

```python
# spliting the data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.iloc[:,-1:], test_size=0.2, random_state=42)
```

```python
print(f'Training data shape: {X_train.shape, y_train.shape}')
print(f'Testing data shape : {X_test.shape, y_test.shape}')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_transformed = scaler.fit_transform(X_train)
X_test_transformed = scaler.transform(X_test)
```

```python
X_train_transformed = pd.DataFrame(X_train_transformed, columns=X_train.columns)
X_test_transformed = pd.DataFrame(X_test_transformed, columns=X_test.columns)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,5))

sns.distplot( data['age'],ax=axes[0],color='green')
plt.title('before scaling')

sns.distplot( X_train['age'],ax=axes[1],color='green')
plt.title('after scaling')

plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_transformed, y_train)
```

```python
actual_y_pred = lr.predict(X_test)
scaled_y_pred = lr_scaled.predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score

print(f'actual    : {accuracy_score(y_test, actual_y_pred)}')
print(f'scaled    : {accuracy_score(y_test, scaled_y_pred)}')
```

```python
from sklearn.tree import DecisionTreeClassifier

dtr = DecisionTreeClassifier()
dtr_scaled = DecisionTreeClassifier()
```

```python
dtr.fit(X_train,y_train)
dtr_scaled.fit(X_train_transformed, y_train)
```

```python
actual_y_pred_dtr = dtr.predict(X_test)
scaled_y_pred_dtr = dtr_scaled.predict(X_test_transformed)
```

```python
print(f'actual    : {accuracy_score(y_test, actual_y_pred_dtr)}')
print(f'scaled    : {accuracy_score(y_test, scaled_y_pred_dtr)}')
```

## 2. Normalization

### 2.1 MinMax Scalling
- MinMax Scaler scaled data between [0-1] only.
<img src='https://pythonsimplified.com/wp-content/uploads/2022/06/MinMaxScaler.jpg?ezimgfmt=rs:372x139/rscb5/ng:webp/ngcb4'>

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaled = MinMaxScaler()

X_train_minmax = pd.DataFrame(min_max_scaled.fit_transform(X_train),columns=X_train.columns)
X_test_minmax = pd.DataFrame(min_max_scaled.transform(X_test), columns=X_test.columns)
```

```python
X_train_minmax.sample(5)
```

```python
X_train_minmax.describe()
```
